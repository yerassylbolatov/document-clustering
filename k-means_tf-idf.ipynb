```python
file_path = "/content/drive/MyDrive/Colab Notebooks/ML course - unik/texts"

def read_file(file):
  f = open(file_path+"/"+file, "r", encoding='utf-8')
  text = f.read()
  f.close()
  return text
```

```python
words_by_texts = []

for file in os.listdir(file_path):
  if file.endswith(".txt"):
    text = read_file(file)
    # print(file)
    words_by_texts.append(text.split())

print("amount of texts:",len(words_by_texts))
```

```python
unique_words = set()

for text in words_by_texts:
  unique_words.update(set(text))

print("number of unique words:",len(unique_words))
```

```python
def cleaning(words):
  stop_words = set(stopwords.words('english'))
  res = set()
  for word in words:
    if word.isalpha():
      lowercase = word.lower()
      if lowercase not in stop_words:
        res.add(lowercase)

  return res

all_words = cleaning(unique_words)
# print(all_words)
print("amount of all cleanining, unique words:",len(all_words))
```

```python
def norm_each_word(word):
  stop_words = set(stopwords.words('english'))
  if word.isalpha():
    lowercase = word.lower()
    if lowercase not in stop_words:
      return lowercase
```

```python
bug_of_words = []
len_of_clean_texts = []


for text in words_by_texts:
  bug_dict = dict.fromkeys(all_words, 0)
  clean_text = cleaning(set(text))
  len_of_clean_texts.append(len(clean_text))
  for word in text:
    key = norm_each_word(word)
    if key in clean_text:
      bug_dict[key]+=1
  bug_of_words.append(bug_dict)


# print(bug_of_words[0])
df = pd.DataFrame(bug_of_words)
df
```

```python
def tf_compute(wordDict, len_of_clean_texts):
  tfDict = {}
  for word, count in wordDict.items():
    tfDict[word] = count / float(len_of_clean_texts)
  return tfDict

tf = []
i = 0
for bugs in bug_of_words:
  tf.append(tf_compute(bugs,len_of_clean_texts[i]))
  i+=1

# print(length_of_clean_texts)

df = pd.DataFrame(tf)
df
```

```python
def idf_compute(documents):
  import math
  N = len(documents)
  print(N)
  
  idfDict = dict.fromkeys(documents[0].keys(), 0)

  # in how many documents each word appears
  for document in documents: 
    for word, val in document.items():
      if val > 0:
        idfDict[word] += 1

  for word, val in idfDict.items():
    idfDict[word] = math.log(N / float(val))
    
  return idfDict

idf = idf_compute(bug_of_words)
# idf
```

```python
tfidf = []

def tfidf_compute(tf, idf):
  tf_idf = {}
  for word, val in tf.items():
    tf_idf[word] = val * idf[word]
  return tf_idf

for i in range(0,len(words_by_texts)):
  tfidf.append(tfidf_compute(tf[i],idf))

X = pd.DataFrame(tfidf)
X

# print(tfidf)
```

```python
def random_centroids(data, k):
  centroids = []
  for i in range(k):
    centroid = data.apply(lambda x: float(x.sample()))
    centroids.append(centroid)
  return pd.concat(centroids, axis=1)
```

```python
centroids = random_centroids(X, 3)
centroids
```

```python
def get_labels(data, centroids):
  distances = centroids.apply(lambda x: np.sqrt(((data - x) ** 2).sum(axis=1)))
  # print("Euclidian distances of each document to each centroid: \n",distances)
  return distances.idxmin(axis=1)
```

```python
labels = get_labels(X, centroids)
print("\nClusters that have minimal distance:")
labels
```

```python
def new_centroids(data, labels, k):
  centroids = data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
  return centroids
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def plot_clusters(X, labels, centroids, iteration):
  pca = PCA(n_components=2)
  data_2d = pca.fit_transform(X)
  centroids_2d = pca.transform(centroids.T)
  clear_output(wait=True)
  plt.title(f'Iteration {iteration}')
  plt.scatter(x=data_2d[:,0], y=data_2d[:,1], c=labels)
  plt.scatter(x=centroids_2d[:,0],y=centroids_2d[:,1])
  plt.show()
```

```python
max_iterations = 10
k = 3

centroids = random_centroids(X, k)
old_centroids = pd.DataFrame()
iteration = 1

while iteration < max_iterations:
  old_centroids = centroids
  # print(labels.value_counts())
  labels = get_labels(X, centroids)
  centroids = new_centroids(X, labels, k)
  # plot_clusters(X, labels, centroids, iteration)
  iteration += 1
```
